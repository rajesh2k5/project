# matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# Bar chart

```python
tesla =['close','volume','open','high','low']
percentage=[270.49,47.87699,264.50,273.8800,262.2400]
```

```python
plt.bar(tesla, percentage,color="purple",edgecolor="black")
```

# Scatter plot

```python
plt.scatter(tesla, percentage,color="green")
```

# Line plot

```python
plt.plot(tesla, percentage, linestyle="solid")
```

# Pie chart

```python
plt.pie(percentage, labels=tesla)
```
