```python
newdataset=TE.drop(['date','volume','high','low'],axis=1)
```

```python
newdataset
```

```python
x=newdataset.iloc[:,0].values.reshape(-1,1)
y=newdataset.iloc[:,1].values.reshape(-1,1)
```

```python
x
```

```python
y
```

```python
pt.scatter(x,y)
pt.xlabel('close')
pt.ylabel('open')
pt.title('Tesla-stock-price')
pt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=0)
```

```python
x_train.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
y_predict=lm.predict(x_test)
```

```python
y_predict
```

```python
pt.scatter(x,y,color='purple')
pt.plot(x_test,y_predict,color='black')
pt.xlabel('close')
pt.ylabel('open')
pt.title('Tesla-stock-price')
pt.show()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
mean_squared_error(y_test,y_predict)
```

```python
mean_absolute_error(y_test,y_predict)
```

```python
r2_score(y_test,y_predict)
```
